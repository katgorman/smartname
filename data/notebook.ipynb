Plotting curves in two-dimensions is nice and all, but Sage is extremely helpful when it comes to plotting (and thus visualizing) surfaces in three-dimensions. We will be plotting surfaces of the form $z = f(x,y)$ and so we first need to tell Sage that our variables will be $x$ and $y$.

```python
# To run this code, either click the "Play" button above or click on this code block and press "Shift + Enter/Return".

x = var('x')
y = var('y')

print('Yippee! x and y will be our variables for the rest of this lab!')
```

To see the syntax of plotting surfaces with Sage, let's do an example. Let's plot $z=2\cos(3x)$ over the rectangle $[-\pi,\pi]\times [-\pi,\pi]$:

```python
# To run this code, either click the "Play" button above or click on this code block and press "Shift + Enter/Return".

plot3d(2*cos(3*x), (x,-pi,pi), (y,-pi,pi))
```

You can use your cursor to move the box containing the surface around to get different views of it. Try it! 

We can also add optional commands to make the surface easier to analyze. Below we will add some gridlines using the _mesh_ variable and some varying colors with the _color_ variable:

```python
# To run this code, either click the "Play" button above or click on this code block and press "Shift + Enter/Return".

plot3d(2*cos(3*x), (x,-pi,pi), (y,-pi,pi), mesh = True)
```
